```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Using MATLAB
plt.plot(x_axis,y_axis)
```

```python
#Using Object oriented interface

fig, ax= plt.subplots()  # Unpacks tuple into 2 variables fig and ax
ax.plot(x_axis,y_axis)
```

```python
#Alternative method to create figure and axis
fig=plt.figure()
ax = fig.add_subplot()

ax.plot(x_axis,y_axis)
```

```python
#Annotate the chart using MATLAB

#Create plot and add label for the legend
plt.plot(x_axis,y_axis,marker="*",color="blue",linewidth=2, label='Boston')

#Create labels for x and Y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 to 45
plt.ylim(0,45)

#Create title
plt.title("PyBer Fare By Month")

#Add legend
plt.legend()

plt.grid()
```

```python
#Annotation using object oriented method
fig=plt.figure()
ax = fig.add_subplot()

ax.plot(x_axis,y_axis,marker="*", color="green",label='Boston')
ax.set_title("PyBer Fare By Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fares($)")
ax.set_ylim(0,45)
ax.legend()
ax.grid()
```

```python
# Vertical Bar Charts
x_axis=["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

```

```python
plt.bar(x_axis,y_axis,color="green",label="Boston")

#Create labels for x and y axid
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Create title
plt.title("PyBer Fare By Month")

#Add legend
plt.legend()
```

```python
#Horizontal bar chart
plt.barh(x_axis,y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()
plt.legend()
plt.title("PyBer Fare By Month")

#Create labels for x and y axid
plt.xlabel("Fare($)")
plt.ylabel("Date")
```

```python
# Vertial Bar chart using object oriented method

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

fig, ax = plt.subplots()
ax.bar(x_axis,y_axis,color="black",label="New York")
ax.legend()
```

```python
#Horizontal bar chart using object oriented method

fig,ax=plt.subplots()
ax.barh(x_axis,y_axis,color="cyan",label="Chicago")
ax.invert_yaxis()
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.legend()
```

```python
#Scatter plot
plt.plot(x_axis,y_axis,'o')
```

```python
plt.scatter(x_axis,y_axis, c="red", label="Chicago")
plt.title("PyBer Fare by Month")
plt.xlabel("Time")
plt.ylabel("Fare($)")
plt.scatter(y_axis,x_axis,c="Red")
plt.gca().invert_yaxis()
plt.legend()
```

```python
#Bubble chart
plt.scatter(x_axis, y_axis, s=y_axis)

#Increase size of the marker
y_axis_larger =[]
for data in y_axis:
    y_axis_larger.append(data*3)
    
#OR

plt.scatter(x_axis,y_axis, [i*3 for i in y_axis])
    
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
#Scatter plot using object oriented method
fig, ax=plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
#Bubble chart using object oriented method

fig,ax=plt.subplots()
ax.scatter(x_axis,y_axis,[i*5 for i in y_axis], color="skyblue",alpha=0.2, edgecolors='black',label="Boston")
ax.legend()
ax.set_xlabel("Time")
ax.set_ylabel("Fare($)")
```

```python
#Pie chart using matlab

plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#explode values for July
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values=(0,0,0,0,0,0,0.3,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,labels=x_axis,explode=explode_values, autopct='%.1f%%',colors=colors)

plt.show()
```

```python
#create pie chart using object oriented method
fig,ax=plt.subplots(figsize=(8,8))
explode_values=[0,0,0,0,0,0,0.2,0,0,0,0,0]
ax.pie(y_axis,labels=x_axis,explode=explode_values,autopct='%.1f%%')
plt.show()
```
